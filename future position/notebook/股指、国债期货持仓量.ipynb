## 按合约跟踪标的分类统计持仓

### 10年期国债期货

```python
t_df = df[df['OptionCode'] == 502]
```

```python
t_pos = t_df.groupby('TradingDay')['OpenInterest'].sum()
```

### 5年期国债期货

```python
tf_df = df[df['OptionCode'] == 501]
```

```python
tf_pos = tf_df.groupby('TradingDay')['OpenInterest'].sum()
```

### 2年期国债期货

```python
ts_df = df[df['OptionCode'] == 503]
```

```python
ts_pos = ts_df.groupby('TradingDay')['OpenInterest'].sum()
```

```python
bond_pos = t_pos.add(tf_pos, fill_value=0).add(ts_pos, fill_value=0)
```

```python
bond_pos.plot(figsize=(16, 9))
```

### 沪深300股指期货

```python
if_df = df[df['OptionCode'] == 3145]
```

```python
if_pos = if_df.groupby('TradingDay')['OpenInterest'].sum()
```

### 上证50股指期货

```python
ih_df = df[df['OptionCode'] == 46]
```

```python
ih_pos = ih_df.groupby('TradingDay')['OpenInterest'].sum()
```

### 中证500股指期货

```python
ic_df = df[df['OptionCode'] == 4978]
```

```python
ic_pos = ic_df.groupby('TradingDay')['OpenInterest'].sum()
```

```python
stock_pos = if_pos.add(ih_pos, fill_value=0).add(ic_pos, fill_value=0)
```

```python
stock_pos.plot(figsize=(16, 9))
```

# 股指期货与国债期货持仓量Z值之差

```python
# 因为2015年牛市的干扰，股债只取2016年之后的数据
stock_pos = stock_pos[stock_pos.index >= '2016-01-01']
bond_pos = bond_pos[bond_pos.index >= '2016-01-01']
```

```python
stock = (stock_pos - stock_pos.mean()) / stock_pos.std()
bond = (bond_pos - bond_pos.mean()) / bond_pos.std()
```

```python
diff = stock - bond
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(stock, label=u'股票', color='blue')
ax.plot(bond, label=u'债券', color='green')
ax.bar(diff.index, diff, color='red')
ax.legend()
```
