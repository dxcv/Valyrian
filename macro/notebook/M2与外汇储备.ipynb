```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(u'../data/M2与外汇储备.xlsx', index_col=0)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df = df.resample('M').last()
```

```python
df.head()
```

# 韩国M2比外储

```python
df[u'韩国M2/外储'] = (df[u'韩国:广义M2'] / df[u'美元兑韩元']) * 1e6 / df[u'韩国:国际储备:外汇储备']
```

```python
df[u'韩国M2/外储'].plot(figsize=(16, 9))
```

# 泰国M2比外储

```python
df[u'泰国M2/外储'] = (df[u'泰国:广义货币(M2)'] / df[u'美元兑泰铢']) / df[u'泰国:国际储备:美元:外汇储备']
```

```python
df[u'泰国M2/外储'].plot(figsize=(16, 9))
```

# 中国M2比外储

```python
df[u'中国M2/外储'] = (df['M2'] / df[u'美元兑人民币元']) / df[u'官方储备资产:外汇储备']
```

```python
df[u'中国M2/外储'].plot(figsize=(16, 9))
```
