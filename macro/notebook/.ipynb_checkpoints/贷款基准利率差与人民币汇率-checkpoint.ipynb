# 中美贷款基准利率之差与人民币兑美元汇率之间的关系

根据利率平价，中美基准利率之差应该与人民币对美元汇率呈正相关

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(u'../data/中美基准利率.xlsx', index_col=0)
```

```python
df.head()
```

```python
df = df.fillna(method='ffill')
```

```python
df['diff'] = df[u'美国联邦基金利率'] - df[u'1年期贷款基准利率']
```

```python
df.plot(figsize=(16, 9))
```

```python
usdcnh = pd.read_excel('../data/USDCNH.xlsx', index_col=0)
```

```python
df = df[df.index >= usdcnh.index[0]]
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(df.index, df['diff'], color='red', label=u'中美基准利率差')
ax.legend()
ax1 = ax.twinx()
ax1.plot(usdcnh.index, usdcnh['USDCNY.IB'], label=u'人民币兑美元汇率')
ax1.legend()
```
