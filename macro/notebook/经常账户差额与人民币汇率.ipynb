# 经常账户差额与人民币汇率的关系

经常账户顺差时，人民币应该升值；经常账户逆差时，人民币应该贬值

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(u'../data/经常账户差额.xlsx', index_col=0)
```

```python
df.head()
```

```python
usdcnh = pd.read_excel('../data/USDCNH.xlsx', index_col=0)
```

```python
df = df[df.index >= usdcnh.index[0]]
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.bar(df.index, df[u'经常账户差额'], color='blue', width=50, label=u'经常账户差额')
ax.legend()
ax1 = ax.twinx()
ax1.plot(usdcnh.index, usdcnh['USDCNY.IB'], color='red', label=u'人民币兑美元汇率')
ax1.legend()
```
