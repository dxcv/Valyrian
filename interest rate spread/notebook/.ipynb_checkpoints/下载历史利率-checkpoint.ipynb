```python
from WindPy import w
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w.start()
```

```python
codes = ['M0017138', # SHIBOR overnigh
         'M0017142', # SHIBOR 3M
]
```

```python
data = w.edb(codes, '2006-10-01', '2018-03-23')
```

```python
df = pd.DataFrame(np.array(data.Data).T, index=data.Times, columns=data.Codes)
```

```python
df.plot(figsize=(16, 9))
```

```python
df.to_excel('../data/shibor.xlsx')
```
