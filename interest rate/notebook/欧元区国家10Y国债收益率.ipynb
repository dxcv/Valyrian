```python
import pandas as pd
import numpy as np
from WindPy import w
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
w.start()
```

```python
codes = 'G0008068,G2700075,G8520070'
```

```python
data = w.edb(codes, '1997-09-01', datetime.date.today())
```

```python
df = pd.DataFrame(np.array(data.Data).T, index=data.Times, columns=[u'德国', u'西班牙', u'希腊'])
```

```python
df = df.fillna(method='ffill')
```

```python
df.tail()
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

lns1 = ax.plot(df[u'德国'], color='blue', label=u'德国')
lns2 = ax.plot(df[u'西班牙'], color='red', label=u'西班牙')
lns3 = ax.plot(df[u'希腊'], color='black', label=u'希腊')

# zero line
ax.axhline(y=0, color='black', linewidth=1)

# Legend
lns = lns1 + lns2 + lns3
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, frameon=False, fontsize=15, loc='upper center', ncol=3, bbox_to_anchor=(0.5, 1))

# Axis
ax.yaxis.tick_right()
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
vals = ax.get_yticks()
ax.set_yticklabels(['{:2.2f}%'.format(x) for x in vals])

# Title
_ = ax.set_title(u'欧元区国家10Y国债收益率', fontsize=25, y=1.02)
fig.savefig(u'../pic/欧元区国家10Y国债收益率.jpg', format='jpg', dpi=200, bbox_inches='tight')
```
