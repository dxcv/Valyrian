```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('../data/BDI.xlsx')
```

```python
df.head()
```

```python
imp = pd.read_excel(u'../data/中国进口金额累计同比.xlsx')
```

```python
imp.head()
```

```python
fig, ax = plt.subplots(figsize=(16, 9))

lns1 = ax.plot(df[df.index <= '2019-01-01'][u'波罗的海干散货指数(BDI)'], color='blue', label='BDI')
lns2 = ax.plot(df[df.index > '2019-01-01'][u'波罗的海干散货指数(BDI)'], color='blue', linestyle='--', label=u'今年BDI')
axx = ax.twinx()
lns3 = axx.plot(imp[u'中国:进口金额:累计同比'], color='red', label=u'中国进口同比增速')

# Legend
lns = lns1 + lns2 + lns3
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, frameon=False, fontsize=15, loc='upper center', ncol=3, bbox_to_anchor=(0.5, 1))

# Axis
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
axx.yaxis.set_tick_params(labelsize=15)
vals = axx.get_yticks()
axx.set_yticklabels(['{:3.0f}%'.format(x) for x in vals])

# Title
_ = ax.set_title(u'波罗的海干散货指数(BDI)', fontsize=25, y=1.05)
```
